```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
percentages = [0,10,20,30,40,50,60,70]
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"],
    'font.size': 15})
results_path = '/nas/home/lcomanducci/xai_src_loc/endtoend_src_loc2/results_perturbation'
```

# T60: 0.15 s SNR: 25 dB

```python
SNR=25
T60=0.15
results_split_path = os.path.join(results_path, 'SNR_' + str(SNR) + '_T60_' + str(T60))
MAE_relevance = np.load(os.path.join(results_split_path,'relevance.npy'))
MAE_random = np.load(os.path.join(results_split_path,'random.npy'))
MAE_energy = np.load(os.path.join(results_split_path,'energy.npy'))
plt.figure(figsize=(7,1.5))
plt.plot(percentages, MAE_relevance,'k-o')
plt.plot(percentages, MAE_random,'k--o')
plt.plot(percentages, MAE_energy,'k:o')
plt.xlabel('$P_{\%} [\%]$')
plt.ylabel('$\mathrm{MAE} [\mathrm{m}]$')
#plt.legend(['$\mathrm{LRP}$','$\mathrm{Random}$','$\mathrm{Absolute}$'],fontsize=10)
plt.savefig('/nas/home/lcomanducci/xai_src_loc/endtoend_src_loc2/plots_icassp/manipulation_SNR_' + str(SNR) + '_T60_' + str(T60)+'.pdf',bbox_inches='tight')

plt.show()
print(MAE_relevance)
```
